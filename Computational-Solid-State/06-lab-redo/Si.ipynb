# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

# VC Relax

```python
# Read the data from prefix.verelax.enth
dcpp, dchh = [], []

with open('Si.vcrelax.dc.enth', 'r') as file:
    for line in file:
        line = line.split()
        dcpp += [int(line[0].split(".")[2])]
        dchh += [float(line[4])]

dcpp, dchh = zip(*sorted(zip(dcpp, dchh)))
dchh = np.array(dchh)


fccpp, fcchh = [], []

with open('Si.vcrelax.fcc.enth', 'r') as file:
    for line in file:
        line = line.split()
        fccpp += [int(line[0].split(".")[2])]
        fcchh += [float(line[4])]

fccpp, fcchh = zip(*sorted(zip(fccpp, fcchh)))
fcchh = np.array(fcchh)



# Plotting
plt.figure(figsize=(10, 4))
plt.title('Enthalpy per Atom vs Pressure')
plt.plot(dcpp, dchh / 2, marker='o', linestyle='-', color='b', label="DC")
plt.plot(fccpp, fcchh, marker='o', linestyle='-', color='r', label="FCC")

plt.xlabel('Pressure (kbar)')
plt.ylabel('Enthalpy per Atom (Ry)')
plt.grid(True)

plt.savefig('Si.vcrelax.png')
plt.legend()
plt.show()
```
