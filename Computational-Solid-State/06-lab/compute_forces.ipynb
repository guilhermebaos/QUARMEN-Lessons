```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from energies.dat
data = np.loadtxt("energy_disp.dat")
displacements = data[:, 0]  # First column: displacements
energies = data[:, 1]       # Second column: total energies

# Compute numerical derivatives using central differences
forces = np.zeros_like(displacements)
for i in range(1, len(displacements) - 1):
    forces[i] = -(energies[i+1] - energies[i-1]) / (displacements[i+1] - displacements[i-1])

# Use forward/backward difference for the boundaries
forces[0] = -(energies[1] - energies[0]) / (displacements[1] - displacements[0])
forces[-1] = -(energies[-1] - energies[-2]) / (displacements[-1] - displacements[-2])

# Plot total energies
plt.figure(figsize=(8, 6))
plt.plot(displacements, energies, '-o', label='Total Energy')
plt.xlabel('Displacement (a.u.)')
plt.ylabel('Total Energy (Ry)')
plt.title('Total Energy vs Displacement')
plt.grid()
plt.legend()
plt.show()

# Plot forces
plt.figure(figsize=(8, 6))
plt.plot(displacements, forces, '-o', label='Force (Finite Difference)', color='red')
plt.xlabel('Displacement (a.u.)')
plt.ylabel('Force (Ry/a.u.)')
plt.title('Force vs Displacement')
plt.grid()
plt.legend()
plt.show()
```
